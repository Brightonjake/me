```python
df = df.drop([576,577], axis=0)
```

```python
def add_centroid(row):
    return row.geometry.centroid

df["centroid"] = df.apply(add_centroid, axis=1)

df[['sa2_maincode_2016', 'Count_home_Census_Nt_0_14_yr', 'geometry', 'centroid']]
```

```python
import shapely
right_here = shapely.geometry.point.Point(151.2299732, -33.9178754)
df["distance_from_UNSW"] = df.geometry.distance(right_here)
```

```python
df.distance_from_UNSW.hist(bins=50);
```

```python
close_burbs = df[df.distance_from_UNSW<0.02]
close_burbs.plot();
```

```python
really_close_burbs = df[df.distance_from_UNSW<0.02]
really_close_burbs.plot()

for idx, row in really_close_burbs.iterrows():
    plt.annotate(s=row.sa2_maincode_2016, 
                 xy=tuple(row.centroid.coords)[0],
                 horizontalalignment='center')
plt.title("Some pretty dodgy suburb names on a map");
```
